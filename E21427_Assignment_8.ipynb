Q. Statistical Mechanics of the Quantum Oscillator

1. The Partition FunctionThe energy eigenvalues are given by:$$E_j = \hbar\omega\left(j + \frac{1}{2}\right) \quad \text{for } j = 0, 1, 2, \dots$$The canonical partition function is:$$Z(T) = \sum_{j=0}^{\infty} e^{-E_j / (k_B T)}$$Substitute $E_j$ into the sum:$$Z(T) = \sum_{j=0}^{\infty} e^{-\frac{\hbar\omega(j + 1/2)}{k_B T}} = \sum_{j=0}^{\infty} e^{-\frac{\hbar\omega}{2k_B T}} \cdot e^{-\frac{j\hbar\omega}{k_B T}}$$We can pull out the constant factor that does not depend on $j$:$$Z(T) = e^{-\frac{\hbar\omega}{2k_B T}} \sum_{j=0}^{\infty} \left(e^{-\frac{\hbar\omega}{k_B T}}\right)^j$$The summation $\sum_{j=0}^{\infty} r^j$ is an infinite geometric series where the common ratio is $r = e^{-\frac{\hbar\omega}{k_B T}}$. Since $T > 0$ and $\hbar\omega > 0$, the ratio satisfies $|r| < 1$, meaning the geometric series is strictly convergent.Using the standard geometric series convergence formula $\sum_{j=0}^{\infty} r^j = \frac{1}{1-r}$, we get:$$Z(T) = e^{-\frac{\hbar\omega}{2k_B T}} \left( \frac{1}{1 - e^{-\frac{\hbar\omega}{k_B T}}} \right) = \frac{e^{-\frac{\hbar\omega}{2k_B T}}}{1 - e^{-\frac{\hbar\omega}{k_B T}}}$$

2. Thermodynamic PotentialsLet's first define $\beta = \frac{1}{k_B T}$. Expressing $Z$ in terms of $\beta$:$$Z = \frac{e^{-\frac{\beta\hbar\omega}{2}}}{1 - e^{-\beta\hbar\omega}}$$Taking the natural logarithm:$$\ln Z = \ln\left(e^{-\frac{\beta\hbar\omega}{2}}\right) - \ln\left(1 - e^{-\beta\hbar\omega}\right) = -\frac{\beta\hbar\omega}{2} - \ln\left(1 - e^{-\beta\hbar\omega}\right)$$

A. Helmholtz Free Energy, $F(T)$ $$F(T) = -k_B T \ln Z = -\frac{1}{\beta} \ln Z$$ $$F(T) = -\frac{1}{\beta} \left[ -\frac{\beta\hbar\omega}{2} - \ln\left(1 - e^{-\beta\hbar\omega}\right) \right] = \frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right)$$

B. Internal Energy, $U(T)$ $$U(T) = -\frac{\partial \ln Z}{\partial \beta} = -\frac{\partial}{\partial \beta} \left[ -\frac{\beta\hbar\omega}{2} - \ln\left(1 - e^{-\beta\hbar\omega}\right) \right]$$$$U(T) = -\left[ -\frac{\hbar\omega}{2} - \frac{1}{1 - e^{-\beta\hbar\omega}} \cdot \left(-e^{-\beta\hbar\omega}\right) \cdot (-\hbar\omega) \right]$$$$U(T) = \frac{\hbar\omega}{2} + \frac{\hbar\omega e^{-\beta\hbar\omega}}{1 - e^{-\beta\hbar\omega}}$$              Multiplying the numerator and denominator of the fraction by $e^{\beta\hbar\omega}$ gives:$$U(T) = \frac{\hbar\omega}{2} + \frac{\hbar\omega}{e^{\beta\hbar\omega} - 1} = \frac{\hbar\omega}{2} + \frac{\hbar\omega}{e^{\frac{\hbar\omega}{k_B T}} - 1}$$

3. EntropyWe find Entropy $S(T)$ using $S = -\left(\frac{\partial F}{\partial T}\right)$. Let's differentiate our expression for $F(T)$:$$F(T) = \frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right)$$Using the product rule on the second term:$$\frac{\partial F}{\partial T} = k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) + k_B T \cdot \frac{1}{1 - e^{-\frac{\hbar\omega}{k_B T}}} \cdot \left(-e^{-\frac{\hbar\omega}{k_B T}}\right) \cdot \left(\frac{\hbar\omega}{k_B T^2}\right)$$$$\frac{\partial F}{\partial T} = k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) - \frac{\hbar\omega}{T} \frac{e^{-\frac{\hbar\omega}{k_B T}}}{1 - e^{-\frac{\hbar\omega}{k_B T}}}$$$$\frac{\partial F}{\partial T} = k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) - \frac{\hbar\omega}{T(e^{\frac{\hbar\omega}{k_B T}} - 1)}$$Therefore, the Entropy $S(T)$ is:$$S(T) = -k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) + \frac{\hbar\omega}{T(e^{\frac{\hbar\omega}{k_B T}} - 1)}$$Verification of $U = F + TS$:$$F + TS = \left[\frac{\hbar\omega}{2} + k_B T \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right)\right] + T \left[-k_B \ln\left(1 - e^{-\frac{\hbar\omega}{k_B T}}\right) + \frac{\hbar\omega}{T(e^{\frac{\hbar\omega}{k_B T}} - 1)}\right]$$The $k_B T \ln(\dots)$ terms perfectly cancel out, leaving:$$F + TS = \frac{\hbar\omega}{2} + \frac{\hbar\omega}{e^{\frac{\hbar\omega}{k_B T}} - 1} = U(T)$$The relation is successfully verified.

Q. The Maximum Entropy Distribution under Mean and Variance Constraints

To prove that the continuous probability distribution that maximizes the differential entropy under the constraints of a fixed mean $\mu$ and fixed variance $\sigma^2$ is the normal (Gaussian) distribution, we use the method of Lagrange Multipliers with calculus of variations.1. Define the Objective and ConstraintsWe want to maximize the differential entropy $h(f)$ of a continuous probability density function $f(x)$ defined over the interval $(-\infty, \infty)$:$$h(f) = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx$$Subject to three constraints:Normalization (Total probability equals 1):$$\int_{-\infty}^{\infty} f(x) \, dx = 1$$Specified Mean ($\mu$):$$\int_{-\infty}^{\infty} x f(x) \, dx = \mu$$Specified Variance ($\sigma^2$):$$\int_{-\infty}^{\infty} (x-\mu)^2 f(x) \, dx = \sigma^2$$2. Set Up the Functional (Lagrange Multipliers)We introduce Lagrange multipliers $\lambda_0$, $\lambda_1$, and $\lambda_2$ for each constraint. The functional to maximize is:$$J[f] = -\int_{-\infty}^{\infty} f(x) \ln f(x) \, dx - \lambda_0 \left( \int_{-\infty}^{\infty} f(x) \, dx - 1 \right) - \lambda_1 \left( \int_{-\infty}^{\infty} x f(x) \, dx - \mu \right) - \lambda_2 \left( \int_{-\infty}^{\infty} (x-\mu)^2 f(x) \, dx - \sigma^2 \right)$$We can collect the terms under a single integral:$$J[f] = \int_{-\infty}^{\infty} \left[ -f(x) \ln f(x) - \lambda_0 f(x) - \lambda_1 x f(x) - \lambda_2 (x-\mu)^2 f(x) \right] dx + \lambda_0 + \lambda_1\mu + \lambda_2\sigma^2$$3. Take the Functional DerivativeTo maximize $J[f]$, we take the functional derivative with respect to $f(x)$ and set it to zero:$$\frac{\delta J}{\delta f(x)} = 0$$Differentiating the integrand with respect to $f(x)$:$$- \ln f(x) - 1 - \lambda_0 - \lambda_1 x - \lambda_2 (x-\mu)^2 = 0$$Solve for $\ln f(x)$:$$\ln f(x) = -(1 + \lambda_0) - \lambda_1 x - \lambda_2 (x-\mu)^2$$Exponentiating both sides gives the functional form of $f(x)$:$$f(x) = \exp\left(-(1 + \lambda_0) - \lambda_1 x - \lambda_2 (x-\mu)^2\right)$$To make evaluating the constants cleaner, we can re-center the linear terms. Let $-(1+\lambda_0) - \lambda_1 x - \lambda_2(x-\mu)^2$ be structurally rewritten into a single completed square of the form $-C_0 - C_2(x-C_1)^2$. Doing this or solving the constraints shows that for the mean to be exactly $\mu$, the linear term center must align with $\mu$, meaning $\lambda_1 = 0$.Thus, the expression simplifies cleanly to:$$f(x) = A \exp\left(-\lambda_2 (x-\mu)^2\right) \quad \text{where } A = e^{-(1+\lambda_0)}$$4. Solve for the Constants Using ConstraintsConstraint 1: NormalizationUsing the standard Gaussian integral $\int_{-\infty}^{\infty} e^{-a(x-b)^2} dx = \sqrt{\frac{\pi}{a}}$:$$\int_{-\infty}^{\infty} A e^{-\lambda_2 (x-\mu)^2} dx = A \sqrt{\frac{\pi}{\lambda_2}} = 1 \implies A = \sqrt{\frac{\lambda_2}{\pi}}$$Constraint 3: Variance$$\int_{-\infty}^{\infty} (x-\mu)^2 f(x) \, dx = \sigma^2$$Substitute $A = \sqrt{\frac{\lambda_2}{\pi}}$:$$\sqrt{\frac{\lambda_2}{\pi}} \int_{-\infty}^{\infty} (x-\mu)^2 e^{-\lambda_2 (x-\mu)^2} dx = \sigma^2$$Using the standard integral identity $\int_{-\infty}^{\infty} y^2 e^{-\alpha y^2} dy = \frac{1}{2\alpha}\sqrt{\frac{\pi}{\alpha}}$ (with $y = x-\mu$ and $\alpha = \lambda_2$):$$\sqrt{\frac{\lambda_2}{\pi}} \left( \frac{1}{2\lambda_2} \sqrt{\frac{\pi}{\lambda_2}} \right) = \sigma^2$$$$\frac{1}{2\lambda_2} = \sigma^2 \implies \lambda_2 = \frac{1}{2\sigma^2}$$Determine the Amplitude ($A$)Now substitute $\lambda_2 = \frac{1}{2\sigma^2}$ back into our expression for $A$:$$A = \sqrt{\frac{\lambda_2}{\pi}} = \sqrt{\frac{1}{2\pi\sigma^2}}$$5. Final ResultSubstitute $A$ and $\lambda_2$ back into the probability density function $f(x)$:$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x-\mu)^2}{2\sigma^2} \right)$$This is precisely the probability density function of a Normal (Gaussian) distribution, $X \sim N(\mu, \sigma^2)$. Therefore, the distribution that maximizes entropy given a fixed mean and variance over the entire real line is the Gaussian distribution.

Q. Asset Pricing and Portfolio Optimization.

To find the risk-neutral probability density function $p(S_T)$ for the future stock price $S_T > 0$ using the Principle of Maximum Entropy (MaxEnt), we solve a constrained optimization problem via calculus of variations and Lagrange multipliers.1. Define the Objective and ConstraintsWe want to maximize the continuous differential entropy of the distribution $p(S_T)$:$$h(p) = -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T$$subject to the following four empirical constraints:Normalization:$$\int_{0}^{\infty} p(S_T) \, dS_T = 1$$Expected Price (Forward Price):$$\mathbb{E}[S_T] = \int_{0}^{\infty} S_T \, p(S_T) \, dS_T = 102$$Expected Squared Price (Volatility):$$\mathbb{E}[S_T^2] = \int_{0}^{\infty} S_T^2 \, p(S_T) \, dS_T = 10,600$$Market Skew (Downside Crash Risk):$$\mathbb{E}[\ln(S_T)] = \int_{0}^{\infty} \ln(S_T) \, p(S_T) \, dS_T = 4.60$$2. Set Up the Lagrangian FunctionalWe introduce the Lagrange multipliers $\lambda_0, \lambda_1, \lambda_2, \text{ and } \lambda_3$ corresponding to each constraint, respectively. The objective functional to maximize is:$$J[p] = -\int_{0}^{\infty} p(S_T) \ln p(S_T) \, dS_T - \lambda_0 \left( \int_{0}^{\infty} p(S_T) \, dS_T - 1 \right) - \lambda_1 \left( \int_{0}^{\infty} S_T \, p(S_T) \, dS_T - 102 \right) - \lambda_2 \left( \int_{0}^{\infty} S_T^2 \, p(S_T) \, dS_T - 10,600 \right) - \lambda_3 \left( \int_{0}^{\infty} \ln(S_T) \, p(S_T) \, dS_T - 4.60 \right)$$Combining the integrals under a single expression:$$J[p] = \int_{0}^{\infty} \left[ -p(S_T) \ln p(S_T) - \lambda_0 p(S_T) - \lambda_1 S_T p(S_T) - \lambda_2 S_T^2 p(S_T) - \lambda_3 \ln(S_T) p(S_T) \right] dS_T + \text{constants}$$3. Take the Functional DerivativeTo maximize $J[p]$, we take its functional derivative with respect to $p(S_T)$ and set it equal to zero:$$\frac{\delta J}{\delta p(S_T)} = 0$$Differentiating the expression inside the integral yields:$$-\ln p(S_T) - 1 - \lambda_0 - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T) = 0$$Now, isolate $\ln p(S_T)$:$$\ln p(S_T) = -(1 + \lambda_0) - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T)$$4. Find the Analytical Form of the DistributionExponentiating both sides to solve for $p(S_T)$:$$p(S_T) = \exp\left(-(1 + \lambda_0) - \lambda_1 S_T - \lambda_2 S_T^2 - \lambda_3 \ln(S_T)\right)$$Using algebraic properties of exponents ($e^{-\lambda_3 \ln(S_T)} = S_T^{-\lambda_3}$), we can factor the expression. Let $A = e^{-(1 + \lambda_0)}$ serve as the normalization constant:$$p(S_T) = A \cdot S_T^{-\lambda_3} \cdot \exp\left(-\lambda_1 S_T - \lambda_2 S_T^2\right) \quad \text{for } S_T > 0$$5. Interpretation of the Distribution FormThe derived Maximum Entropy PDF belongs to a generalized family of distributions:The $S_T^{-\lambda_3}$ factor directly represents the power-law correction that creates the asymmetrical heavy downside tails ("fat tails"), matching the market's crash-risk premium ($\mathbb{E}[\ln(S_T)] = 4.60$).The $\exp(-\lambda_1 S_T - \lambda_2 S_T^2)$ factor acts as a Gaussian-like damping component ensuring the distribution remains well-behaved and integrates perfectly to $1$ over large values.This distribution represents the mathematically least-biased description of Asset X's future price given only the observed options market pricing constraints, without artificially imposing a rigid classical log-normal structure. (Numerically, the parameters $A, \lambda_1, \lambda_2, \lambda_3$ are determined by solving the system of four nonlinear integration constraint equations matching $1, 102, 10,600, \text{ and } 4.60$.)
