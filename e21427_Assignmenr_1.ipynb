iii) The two models define “random point” differently.

First Model: Uniform Over Area

Every equal area region has the same probability.

The outer rings of a disk contain more area than the inner rings, so points are naturally more likely to fall near the boundary.

Second Model: Uniform Radius

The radius itself is chosen uniformly.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create grid for joint density
t = np.linspace(-4, 4, 200)
p = np.linspace(-8, 8, 200)

T, P = np.meshgrid(t, p)

# Joint density
Z = (1/(4*np.pi)) * np.exp(-(T**2)/2 - (P**2)/8)

# Plot contour of joint density
plt.figure(figsize=(8,6))
plt.contourf(T, P, Z, levels=30)
plt.colorbar(label='Density')

plt.xlabel('Temperature Deviation (T)')
plt.ylabel('Pressure Deviation (P)')
plt.title('Joint Density of Y = (T,P)')

plt.show()

# Marginal density of X
x = np.linspace(-4, 4, 400)
fx = norm.pdf(x, 0, 1)

plt.figure(figsize=(8,5))
plt.plot(x, fx)

plt.xlabel('Temperature Deviation')
plt.ylabel('Density')
plt.title('Marginal Density of X')

plt.show()

# Sigma-algebra illustration
plt.figure(figsize=(8,4))

plt.axvline(x=-1, linestyle='--')
plt.axvline(x=1, linestyle='--')

plt.text(0, 0.5, r'$\sigma(X)$ depends only on vertical strips',
         ha='center')

plt.xlim(-3,3)
plt.ylim(0,1)

plt.title(r'Illustration of $\sigma(X)$ being coarser than $\sigma(Y)$')

plt.show()
```
