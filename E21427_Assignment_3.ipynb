We start from the joint Gaussian density of the partitioned vector:

$$
\mathbf{Z} =
\begin{bmatrix}
\mathbf{X} \\
\mathbf{Y}
\end{bmatrix},
\quad
\mathbf{z} =
\begin{bmatrix}
\mathbf{x} \\
\mathbf{y}
\end{bmatrix}.
$$


1.
Start from the joint density

$$
f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})
\propto
\exp\left(
-\frac{1}{2}
\begin{bmatrix}
\mathbf{x}-\mu_x \\
\mathbf{y}-\mu_y
\end{bmatrix}^T
\Sigma^{-1}
\begin{bmatrix}
\mathbf{x}-\mu_x \\
\mathbf{y}-\mu_y
\end{bmatrix}
\right)
$$

To form the conditional density:

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}|\mathbf{y})
=
\frac{f_{\mathbf{X},\mathbf{Y}}(\mathbf{x},\mathbf{y})}
{f_{\mathbf{Y}}(\mathbf{y})}
$$

So we only need to isolate all terms involving $\mathbf{x}$, treating $\mathbf{y}$ as fixed.

2.
Use block matrix inverse structure (key step)

Let the precision matrix be partitioned as:

$$
\Sigma^{-1} =
\begin{bmatrix}
A & B \\
B^T & C
\end{bmatrix}.
$$

Then the quadratic form expands as:

$$
(\mathbf{x}-\mu_x)^T A (\mathbf{x}-\mu_x)
+ 2(\mathbf{x}-\mu_x)^T B (\mathbf{y}-\mu_y)
+ (\mathbf{y}-\mu_y)^T C (\mathbf{y}-\mu_y).
$$

Only the first two terms depend on $\mathbf{x}$.


3.
Complete the square in $\mathbf{x}$

Group terms:

$$
(\mathbf{x}-\mu_x)^T A (\mathbf{x}-\mu_x)
+ 2(\mathbf{x}-\mu_x)^T B (\mathbf{y}-\mu_y).
$$

This can be rewritten as:

$$
(\mathbf{x}-\mu_x - A^{-1}B(\mathbf{y}-\mu_y))^T
A
(\mathbf{x}-\mu_x - A^{-1}B(\mathbf{y}-\mu_y))
+ \text{(terms independent of } \mathbf{x}).
$$

Thus the conditional mean is:

$$
\mu^* = \mu_x + A^{-1}B(\mathbf{y}-\mu_y).
$$

Now we identify the standard covariance identity for Gaussians:

$$
A^{-1} = \Sigma_{X|Y}
= \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}.
$$

Also:

$$
A^{-1}B = \Sigma_{XY}\Sigma_{YY}^{-1}.
$$

So:

$$
\mu^* = \mu_x + \Sigma_{XY}\Sigma_{YY}^{-1}(\mathbf{y}-\mu_y).
$$


4.
Conditional covariance

From the completed square, the quadratic term in $\mathbf{x}$ has precision matrix $A = \Sigma_{X|Y}^{-1}$, hence:

$$
\Sigma_{X|Y}
=
\Sigma_{XX}
-
\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}.
$$

This is the Schur complement of $\Sigma_{YY}$ in $\Sigma$.


5.
Final conditional density

$$
f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x}|\mathbf{y})
=
\frac{1}{\sqrt{(2\pi)^n |\Sigma_{X|Y}|}}
\exp\left(
-\frac{1}{2}
(\mathbf{x}-\mu^*)^T
\Sigma_{X|Y}^{-1}
(\mathbf{x}-\mu^*)
\right),
$$

where:

$$
\mu^* = \mu_x + \Sigma_{XY}\Sigma_{YY}^{-1}(\mathbf{y}-\mu_y),
\quad
\Sigma_{X|Y}
=
\Sigma_{XX}
-
\Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX}.
$$

2.
Why $X$ is not $\sigma(Y)$-measurable

$\sigma(Y)$-measurable means: once you know $Y$, $X$ is a deterministic function of it.

But here:

$$
Y = X + Q
$$

Both $X$ and $Q$ are random and independent. So the same observed $Y$ can come from infinitely many decompositions:

- high $X$, low $Q$
- medium $X$, medium $Q$
- low $X$, high $Q$

So observing $Y=5$ does not identify a unique $X$. The system is underdetermined.


Interpretation

- $Y$ conflates signal ($X$) and noise ($Q$)
- Without observing $Q$, you cannot separate them

Thus:

$$
X \notin \sigma(Y).
$$


3.
Conditional expectation $E[X \mid Y]$

For jointly Gaussian variables:

$$
E[X \mid Y=y]
=
\mu_X + \frac{\mathrm{Cov}(X,Y)}{\mathrm{Var}(Y)}(y-\mu_Y).
$$

Substitute:

$$
\mu_X = 0,\quad \mu_Y = 0,\quad \mathrm{Cov}(X,Y)=1,\quad \mathrm{Var}(Y)=5.
$$

So:

$$
E[X \mid Y=y] = \frac{1}{5}y.
$$

Final:

$$
E[X \mid Y] = \frac{1}{5}Y.
$$

Signal-to-total variance ratio:

$$
\frac{\mathrm{Cov}(X,Y)}{\mathrm{Var}(Y)} = \frac{1}{5}.
$$

So only 20\% of the observed power fluctuation is attributed to temperature.

7.
Why $Y$ is a bad estimator of $X$

The manager’s estimator is:

$$
\tilde{X} = Y = X + Q.
$$


Bias

$$
E[Y \mid X] = X
$$

So $Y$ is unbiased for $X$ in expectation.

Also:

$$
E[Y] = 0 = E[X].
$$

So the mean is fine, but the issue is variance contamination.


More importantly:

$$
E[Y \mid Y] = Y \neq E[X \mid Y].
$$


Key issue: noise inflation

Using $Y$ as an estimator gives:

$$
\mathrm{MSE}(Y \to X) = \mathrm{Var}(Q) = 4.
$$

But the optimal estimator is:

$$
E[X \mid Y] = \frac{1}{5}Y.
$$

This shrinks because only a fraction of $Y$ is signal; the rest is independent noise.


Why shrinkage occurs

$$
Y = \underbrace{X}_{\text{signal}} + \underbrace{Q}_{\text{noise}}.
$$

Since $Q$ inflates magnitude symmetrically, large $|Y|$ is often partly noise. So Bayesian conditioning pulls estimates toward:

$$
E[X] = 0.
$$


This is the classic regression-to-the-mean effect:

- extreme observations are partially noise-driven  
- optimal estimator dampens them
