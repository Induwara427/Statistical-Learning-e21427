```python
# Constructing the probability space (ő©, F, P)
# for X ~ Beta(2,18)

# Sample space ő©
Omega = "[0,1]"

# Sigma-algebra F
F = "Borel sigma-algebra on [0,1]"

# Beta parameters
alpha = 2
beta = 18

# Define Beta function using math library
import math

def beta_function(a, b):
    return (math.gamma(a) * math.gamma(b)) / math.gamma(a + b)

B = beta_function(alpha, beta)

# Display probability measure definition
print("Probability Space (ő©, F, P)\n")

print("ő© =", Omega)
print("F =", F)

print("\nBeta parameters:")
print("alpha =", alpha)
print("beta =", beta)

print("\nBeta function B(2,18) =", B)

print("\nProbability density function:")
print("f(x) = (1/B(2,18)) * x^(1) * (1-x)^(17),  0 < x < 1")
```

```python
# Probability Density Function of X
# X ~ Beta(2,18)

import sympy as sp

# Define symbols
x = sp.symbols('x', positive=True)

# Parameters
alpha = 2
beta = 18

# Define Beta function
B = sp.beta(alpha, beta)

# Define the PDF
f_x = (1 / B) * (x**(alpha - 1)) * ((1 - x)**(beta - 1))

# Simplify expression
f_x_simplified = sp.simplify(f_x)

# Print the PDF
print("Probability Density Function of X:\n")
sp.pprint(f_x_simplified)

print("\nFor 0 < x < 1")
```

```python
# Verify that f(x) is a valid probability density function
# for X ~ Beta(2,18)

import sympy as sp

# Define variable
x = sp.symbols('x', real=True)

# Parameters
alpha = 2
beta = 18

# Define Beta function
B = sp.beta(alpha, beta)

# Define the density function
f = (1 / B) * (x**(alpha - 1)) * ((1 - x)**(beta - 1))

# Step 1: Check non-negativity
print("Density function f(x):")
sp.pprint(f)

print("\nFor 0 < x < 1, f(x) >= 0")

# Step 2: Verify total integral equals 1
integral = sp.integrate(f, (x, 0, 1))

print("\nIntegral of f(x) from 0 to 1:")
sp.pprint(sp.simplify(integral))

# Final verification
if sp.simplify(integral) == 1:
    print("\nVerified: f(x) is a valid probability density function.")
else:
    print("\nVerification failed.")
```

```python
# Compute the variance Var(X)
# for X ~ Beta(2,18)

import sympy as sp

# Parameters
alpha = 2
beta = 18

# Variance formula for Beta distribution
variance = (alpha * beta) / (((alpha + beta)**2) * (alpha + beta + 1))

# Simplify result
variance_simplified = sp.simplify(variance)

# Print result
print("Variance Var(X):")
sp.pprint(variance_simplified)

print("\nDecimal value:")
print(float(variance_simplified))
```

```python
# Compute P(X <= 0.10)
# for X ~ Beta(2,18)

import sympy as sp

# Define variable
x = sp.symbols('x', real=True)

# Parameters
alpha = 2
beta = 18

# Define Beta function
B = sp.beta(alpha, beta)

# Define the PDF
f = (1 / B) * (x**(alpha - 1)) * ((1 - x)**(beta - 1))

# Compute probability P(X <= 0.10)
probability = sp.integrate(f, (x, 0, 0.10))

# Simplify result
probability_simplified = sp.N(probability)

# Print results
print("P(X <= 0.10) =")
sp.pprint(probability_simplified)
```

```python
import sympy as sp

# Define variable
x = sp.Symbol('x')

# Parameters
a, b, c = 4, 10, 6

# PDF pieces
f_left = (x - a) / ((b - a) * (c - a))     # 4 to 6
f_right = (b - x) / ((b - a) * (b - c))    # 6 to 10

# Compute split integrals
part1 = sp.integrate(f_left, (x, 5, 6))
part2 = sp.integrate(f_right, (x, 6, 8))

# Total probability
prob = part1 + part2

print("P(5 ‚Č§ X ‚Č§ 8):", prob)
print("Decimal:", float(prob))
```

```python
import numpy as np
from scipy.special import beta

# Parameters
a = 8
b = 20
c = 11

# Shape parameters
alpha = 1 + 4 * ((c - a) / (b - a))
beta_param = 1 + 4 * ((b - c) / (b - a))

# PERT PDF function
def pert_pdf(x):
    if a <= x <= b:
        numerator = (x - a)**(alpha - 1) * (b - x)**(beta_param - 1)
        denominator = beta(alpha, beta_param) * (b - a)**(alpha + beta_param - 1)
        return numerator / denominator
    else:
        return 0

# Example evaluation
x_values = np.linspace(a, b, 5)
pdf_values = [pert_pdf(x) for x in x_values]

for x, f in zip(x_values, pdf_values):
    print(f"x = {x:.2f}, f(x) = {f:.6f}")
```

```python
import numpy as np
from scipy.integrate import quad
from scipy.special import beta

# Parameters
a = 8
b = 20
c = 11

# Shape parameters
alpha = 1 + 4 * ((c - a) / (b - a))
beta_param = 1 + 4 * ((b - c) / (b - a))

# PERT PDF
def pert_pdf(x):
    if a <= x <= b:
        numerator = (x - a)**(alpha - 1) * (b - x)**(beta_param - 1)
        denominator = beta(alpha, beta_param) * (b - a)**(alpha + beta_param - 1)
        return numerator / denominator
    else:
        return 0

# 1. Theoretical expectation
E_theoretical = (a + 4*c + b) / 6

# 2. Numerical expectation via integration
E_numerical, error = quad(lambda x: x * pert_pdf(x), a, b)

print("Theoretical E[X]:", E_theoretical)
print("Numerical E[X]:", E_numerical)
print("Integration error estimate:", error)
```
